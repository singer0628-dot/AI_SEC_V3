```python
#데이터셋 준비(데이터셋의 속성)
#CRIM: 도시별 1인당 범죄율
#ZN: 25,000 제곱피트 초과 부지로 구획된 주거용 토지의 비율
#INDUS: 도시별 비소매(non-retail) 상업 지역 면적 비율
#CHAS: 찰스강 인접 여부 더미 변수(강과 경계를 이루면 1, 아니면 0)
#NOX: 질소산화물 농도 (1천만 분의 1 단위) - 수질
#RM: 주택 한 채당 평균 방 개수
#AGE: 1940년 이전에 건축된 자가 점유 주택의 비율-노후주택비율
#DIS: 보스턴 내 5개 주요 고용 중심지까지의 가중 거리-취업활성도
#RAD: 방사형 고속도로 접근성 지수 - 교통의 편의성
#TAX: 재산세율 (재산가치 $10,000당) - 재산세
#PTRATIO: 도시별 학생-교사 비율 - 교육의 활성도
#B: 1000 × (Bk − 0.63)²(Bk는 도시별 흑인 인구 비율)
#LSTAT: 하위 계층 인구 비율(%)
#MEDV: 자가 점유 주택의 중위 가격 (단위: 천 달러, $1000) - 정답레이블
```

```python
import numpy as np
# 상관관계 분석후 상관인덱스 리스트와 상관데이터만 추출
index_list = np.array([0,2,4,5,7,11,12])
x_dtrain = x_train[:,index_list]
print(x_dtrain[0])
print(x_train[0])
x_dtest = x_test[:,index_list]
print(x_dtrain.shape)
print(y_train.shape)
print(x_dtest.shape)
print(y_test.shape)
```

```python
# 3. 데이터 스케일링
means=x_dtrain.mean(axis=0)
stds = x_dtrain.std(axis=0)
print("7개 필드의 각각 평균값:\n",means)
print("7개 필드의 각각 표준편차:\n",stds)
```

```python
x_dtrain = (x_dtrain-means)/stds
x_dtest = (x_dtest-means)/stds
print(x_dtrain.shape)
print(x_dtest.shape)
print(x_dtrain[0])
print(x_dtest[0])
#정규화가 잘 수행되었는지 평균과 표준편차 출력
print(x_dtrain.mean(axis=0))
print(x_dtrain.std(axis=0))
print(x_dtest.mean(axis=0))
print(x_dtest.std(axis=0))
```

```python
from tensorflow.keras import Sequential,Input
from tensorflow.keras.layers import Dense
```

```python
#4. 모델 구성
print(x_dtrain.shape)
model = Sequential()
model.add(Input((7,)))
model.add(Dense(1,activation="relu"))
```

```python
#5. 모델 훈련
import random
random.seed(111)
np.random.seed(111)
tf.random.set_seed(111)
model.compile(loss="mse",optimizer="SGD",metrics=["mae"])
res = model.fit(x_dtrain,y_train,epochs=52,validation_split=0.2,batch_size=len(x_dtrain)//10)
```

```python
print(res.history.keys())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.plot(res.history["loss"],label="Train Loss")
plt.plot(res.history["val_loss"],label="Valid Loss")
plt.legend()
plt.title("Losses")
plt.subplot(1,2,2)
plt.plot(res.history["mae"],label="Train MAE")
plt.plot(res.history["val_mae"],label="Valid MAE")
plt.legend()
plt.title("Validation")
plt.show()
```

```python
res = model.evaluate(x_dtest,y_test)
#선형회귀는 평가의 의미가 무의미하다.
print(res)
```

```python
y_pred = model.predict(x_dtest)
```

```python
print(y_pred.shape)
print(y_test.shape)
y_true = y_test.reshape((len(y_test),1))
print(y_pred.shape)
print(y_true.shape)
```

```python
y_gap=1-(y_true/y_pred)
#y_gap 절대값으로 변경
y_gap=np.abs(y_gap)
# 0.01
test = y_gap<=0.01
print(f"오차율 1% 이내인 수량:",test.sum(),\
      f" 총수량대비:{test.sum()/len(y_true)*100:.2f}%")
```

```python
print(f"오차율 평균 : {y_gap.mean()*100:.2f}%")
```
